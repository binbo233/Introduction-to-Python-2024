# 练习11.3 雇员

```python
class Employee:
    """员工的信息"""
    
    def __init__(self, first, last, salary):
        """存储员工的名，姓，年薪"""
        self.first = first
        self.last = last
        self.salary = salary
        
    def give_raise(self, extra=''):
        if extra:
            self.salary += 5000 + int(extra)
        else:
            self.salary += 5000
            
def test_give_default_raise():
    """测试是否加了固定的5000美元"""
    employee_1 = Employee('ziqi', 'Gong', 6000)
    employee_1.give_raise()
    assert employee_1.salary == 11000
    
def test_give_custom_raise():
    """测试是否加了定制的薪水"""
    employee_1 = Employee('ziqi', 'Gong', 6000)
    employee_1.give_raise(1000)
    assert employee_1.salary == 12000
    
#结果
PS C:\Users\binbo233\Desktop\test> python -m pytest
===================================================================================================================================================================== test session starts ======================================================================================================================================================================
platform win32 -- Python 3.11.9, pytest-8.2.1, pluggy-1.5.0
rootdir: C:\Users\binbo233\Desktop\test
collected 2 items

test_name_function.py ..                                                                                                                                                                                                                                                                                                                                  [100%] 

====================================================================================================================================================================== 2 passed in 0.02s ======================================================================================================================================================================= 
```
